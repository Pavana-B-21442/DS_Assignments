```python
X=data.iloc[:,:3]
y=data["TEY"]
```

```python
X1=X.values
```

```python
y1=y.values
```

```python
def plot(model_fit):
    plt.plot(model_fit.history['loss'])
    plt.plot(model_fit.history['val_loss'])
    plt.title('model error')
    plt.ylabel('mse')
    plt.xlabel('epoch')
    plt.legend(['train', 'test'], loc='upper left')
    plt.show()
```

```python
#Building model
### We use relu as activation function n input and hidden layers and linear as activation function in the output layer.
### We use mean squared error as the loss function
model1=Sequential()
model1.add(Dense(3,  activation='relu'))
model1.add(Dense(3,  activation='relu'))
model1.add(Dense(1, activation='linear'))
```

```python
model1.compile(loss='mean_squared_error', optimizer='adam')
```

```python
model_fit1=model1.fit(X1, y1, validation_split=0.33, epochs=250, batch_size=10)
```

```python
model1.evaluate(X1,y1)
```

```python
pred=model1.predict(X1)
pred
```

```python
plot(model_fit1)
```

```python
### decreasing the number of nodes in hidden layer
model2=Sequential()
model2.add(Dense(3,  activation='relu'))
model2.add(Dense(2,  activation='relu'))
model2.add(Dense(1, activation='linear'))
model2.compile(loss='mean_squared_error', optimizer='adam')
model_fit2=model2.fit(X1, y1, validation_split=0.33, epochs=250, batch_size=10)
```

```python
model2.evaluate(X1,y1)
```

```python
plot(model_fit2)
```

```python
### Adding an additional hidden layer to the model
model3=Sequential()
model3.add(Dense(3,  activation='relu'))
model3.add(Dense(3,  activation='relu'))
model3.add(Dense(3,  activation='relu'))
model3.add(Dense(1, activation='linear'))
model3.compile(loss='mean_squared_error', optimizer='adam')
model_fit3=model3.fit(X1, y1, validation_split=0.33, epochs=250, batch_size=10)

```

```python
model3.evaluate(X1,y1)
```

```python
plot(model_fit3)
```

### model1 is computationally simpler with good mse , hence model1 is taken as final model
